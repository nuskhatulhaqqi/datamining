# NAIVE BAYES

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn import tree
import seaborn as sns
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

url = 'https://raw.githubusercontent.com/nuskhatulhaqqi/data_mining/main/credit_score.csv'

data = pd.read_csv(url)
data.drop(columns="Unnamed: 0",inplace=True)
data
```

```python
data_y=pd.get_dummies(data['rata_rata_overdue'])
data_y.head()
```

```python
data_join = data.join(data_y)
data_join.drop(columns="rata_rata_overdue",inplace=True)
```

```python
data_x=pd.get_dummies(data['kpr_aktif'])
data_x.head()
```

```python
data_fix = data_join.join(data_x)
data_fix.drop(columns="kpr_aktif",inplace=True)
data_fix
```

```python
data_risk_rating = pd.DataFrame(data_fix, columns = ['risk_rating'])
data_risk_rating.head()
```

```python
data_join_new = data_fix.join(data_risk_rating)
data_join_new
```

```python
from sklearn.preprocessing import MinMaxScaler
df_for_minmax_scaler=pd.DataFrame(data_join_new, columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
df_for_minmax_scaler.to_numpy()
scaler = MinMaxScaler()
df_hasil_minmax_scaler=scaler.fit_transform(df_for_minmax_scaler)
```

```python
df_hasil_minmax_scaler = pd.DataFrame(df_hasil_minmax_scaler,columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
df_hasil_minmax_scaler
```

```python
df_drop_column_for_minmaxscaler=data_join_new.drop(['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'], axis=1)
df_drop_column_for_minmaxscaler
```

```python
data_new3 = pd.concat([df_drop_column_for_minmaxscaler,df_hasil_minmax_scaler], axis=1)
data_new3
```

```python
df_new3_without_risk_rating=data_new3.drop(['risk_rating'], axis=1)
df_new3_without_risk_rating
```

```python
data_new4 = pd.concat([df_new3_without_risk_rating,data_risk_rating], axis=1)
data_new4
```

# NAIVE BAYES CLASSIFIER

```python
from sklearn.metrics import make_scorer, accuracy_score,precision_score
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold,train_test_split,cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X=data_new4.iloc[:,1:10].values
y=data_new4.iloc[:,11].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, y_train)
Y_pred = gaussian.predict(X_test) 
accuracy_nb=round(accuracy_score(y_test,Y_pred)* 100, 2)
acc_gaussian = round(gaussian.score(X_train, y_train) * 100, 2)

cm = confusion_matrix(y_test, Y_pred)
accuracy = accuracy_score(y_test,Y_pred)
precision =precision_score(y_test, Y_pred,average='micro')
recall =  recall_score(y_test, Y_pred,average='micro')
f1 = f1_score(y_test,Y_pred,average='micro')
print('Confusion matrix for Naive Bayes\n',cm)
print('accuracy_Naive Bayes: %.3f' %accuracy)
print('precision_Naive Bayes: %.3f' %precision)
print('recall_Naive Bayes: %.3f' %recall)
print('f1-score_Naive Bayes : %.3f' %f1)
```
