# Materi
* Atribut biner hanya memiliki dua status: 0 dan 1 Contoh atribut gender menggambarkan seorang siswa, misalnya, 1 menunjukkan bahwa siswa laki-laki, sedangkan 0 menunjukkan siswa perempuan. 
* Memperlakukan atribut biner sebagai atribut numerik tidak diperkenankan. Oleh karena itu, metode khusus untuk data biner diperlukan untuk membedakan komputasi.
* untuk yang symmetric 
$$
\frac{r + s}{q + r + s + t}\
$$
* untuk yang asymmetric
$$
\frac{r + s}{q + r + s}\
$$

```python
# kode CertificationCourse
kode_CertificationCourse_untuk_cherbourg = "No"
kode_CertificationCourse_untuk_southampton = "Yes"
# kode Gender
kode_Gender_untuk_cherbourg = "Female"
kode_Gender_untuk_southampton = "Male"
# nilai biner
biner_satu = 1
biner_nol = 0

# fungsi untuk merubah isi kolom jadi biner
def ubah_kode_CertificationCourse_menjadi_biner(CertificationCourse):
    return biner_satu if CertificationCourse == kode_CertificationCourse_untuk_southampton else biner_nol

def ubah_kode_Gender_menjadi_biner(Gender):
    return biner_satu if Gender == kode_Gender_untuk_southampton else biner_nol
```

```python
# mengubah semua data 
dp["CertificationCourse"] = dp["CertificationCourse"].apply(ubah_kode_CertificationCourse_menjadi_biner)

dp["Gender"] = dp["Gender"].apply(ubah_kode_Gender_menjadi_biner)

#menampilkan data 
dp[["CertificationCourse","Gender"]].head(5)
```

```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(dp, idx, series):
    return dp.loc[(idx - DECREMENT_BY_ONE), series]
```

```python
def get_dissimilarity_dataset(dp, series_index = [], series = []):
    first_series = get_series(dp, series_index[0], series)
    second_series = get_series(dp, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
def count_contingency_value(dp, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = dp.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(dp.loc[(start_index - DECREMENT_BY_ONE):(last_index - DECREMENT_BY_ONE), dp.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
# d(1,2)
d_1_2 = get_dissimilarity_dataset(dp, [1,2], ["CertificationCourse", "Gender"])
d_1_2
```

```python
a = count_contingency_value(d_1_2, 1, 2)
a
```

```python
b = count_contingency_value(d_1_3, 1, 3)
b
```
